```python
import sys
import os
import torch
import pickle
# with open('your_file.pkl', 'rb') as file:
#     loaded_data = pickle.load(file)
```

```python
def load_all_pkl_files(directory):
    all_data = {}

    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith('.pkl'):
                file_path = os.path.join(root, file)
                with open(file_path, 'rb') as f:
                    data = pickle.load(f)
                    all_data[str(file).split('.')[0]] = data
    return all_data
```

```python
def get_rank(tensor):
    _, indices = torch.sort(tensor)
    rank = torch.zeros_like(tensor, dtype=torch.float)
    rank[indices] = torch.arange(1, len(tensor) + 1, dtype=torch.float)
    return rank

def spearman_coefficient(tensor1, tensor2):
    rank1 = get_rank(tensor1)
    rank2 = get_rank(tensor2)
    
    n = len(tensor1)
    
    diff_rank1 = rank1 - torch.mean(rank1)
    diff_rank2 = rank2 - torch.mean(rank2)
    
    numerator = torch.sum(diff_rank1 * diff_rank2)
    denominator = torch.sqrt(torch.sum(diff_rank1 ** 2) * torch.sum(diff_rank2 ** 2))
    
    return numerator / denominator
```

```python
PATH="/data/wyh/SSF/Analysis/saver/200"
all_data = load_all_pkl_files(PATH)
ks=list(all_data.keys())
# sort ks
ks.sort()
for k in ks:
    print(type(k),type(all_data[k]),type(all_data[k][0]),len(all_data[k][0]),list(all_data[k][0].keys())[0])
# for k in ks:
    # print(k,type(k))
for k in all_data[str(1)][1].keys():
    print(k)
```

```python
lks=list(all_data[str(1)][0].keys())
main=str(5)
ks=[int(k) for k in ks]
ks.sort()
print(" ".join([str(k) for k in ks]))
for i,lk in enumerate(lks):
    spes=[]
    for k in ks:
        k=str(k)
        spe=spearman_coefficient(torch.tensor(all_data[main][0][lk]),torch.tensor(all_data[k][0][lk]))
        spes.append(float(spe))

    print(i," ".join(["%.2f" % spe for spe in spes]))
```
